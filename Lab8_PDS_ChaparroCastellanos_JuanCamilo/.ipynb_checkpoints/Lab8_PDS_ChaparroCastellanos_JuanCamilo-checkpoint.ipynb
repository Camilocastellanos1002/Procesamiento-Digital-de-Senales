#### 1.3) Extraccion de Segmentos

```python
def extraer_ventanas(signal, size, fs):
    #Tamano de paso
    step=int(0.010*fs)
    n_seg = int((len(signal) - size) / step)
    # extraer segmentos
    windows = [signal[i * step : i * step + size]
    for i in range(n_seg)]
    # stack (cada fila es una ventana)
    return np.vstack(windows)
```

#### para este punto seleccionamos la señal #1 y seleccionamos el segmento de ventana con 40ms

```python
size1= (sample_senal1*0.04)/time1
print('size: ',size1)
```

```python
segmento=extraer_ventanas(senal1,int(size1),fs1)
```

#### 1.4) aplicacion de ventanas
#### 1.4.1) consulta ventana hamming

#### * Una ventana Hamming es una función de ventana comúnmente utilizada en el procesamiento de señales para reducir el efecto de las discontinuidades en el borde de una señal. Se aplica multiplicando punto a punto la señal original por los valores de la ventana Hamming. Esto suaviza los extremos de la señal y reduce la fuga espectral cuando se realiza una transformada de Fourier en la señal.


#### La fórmula es: 𝑤(𝑛)=0.54−0.46cos(2𝜋𝑛/𝑁−1)
#### donde : 
#### * n es el índice de la muestra y 
#### * 𝑁 es el tamaño de la ventana.

#### * para realizar la ventana hamming se debe utilizar el mismo tamaño de la ventana original, en este caso la ventana de los segmentos durante los 40 ms

```python
ventanaH=np.hamming(size1)
plt.plot(ventanaH)
plt.title('ventana hamming')
plt.xlabel('muestras',fontsize=15) 
plt.ylabel('amplitud',fontsize=15)
plt.grid()
```

#### 1.4.2) Aplique la ventana a los segmentos

```python
ventan_porseg=segmento*ventanaH
plt.plot(ventan_porseg);
plt.grid()
plt.title('ventana hamming * Señal')
plt.xlabel('muestras',fontsize=15) 
plt.ylabel('amplitud',fontsize=15)
```
