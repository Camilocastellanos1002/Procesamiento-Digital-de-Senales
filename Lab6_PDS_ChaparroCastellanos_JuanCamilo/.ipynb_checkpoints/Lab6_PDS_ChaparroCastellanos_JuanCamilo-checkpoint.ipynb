### 2 Generacion de señales

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import IPython

    
#variables
C=8
N=(30+C*2) #periodo de la señal
A=2*(C+1) #amplitud componente positiva
B=-2*(C+1) #amplitud componente negativa

#definicion de la señal x(n)
def x(n, N):
    if n == 0:
        return 0
    elif -N/2 < n < 0:
        return A
    elif 0 < n < N/2:
        return B
    else:
        return 0

print(C)
print(N)
print(A)
print(B)
```

#### 2.1 grafica

```python

# vector tiempo de n
n_values = np.arange(-N/2, N/2 + 1)

# Calcular los valores de la función para cada valor de n
x_values = [x(n, N) for n in n_values]

# Graficar la función
plt.stem(n_values, x_values, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Gráfico de la función x(n)')
plt.grid(True)
plt.show()
```

#### la señal es de caracter par ya que es simetrica alrrededor del eje vertical en n=0, es decir que x(n<0)= 18 mientras que para x(n>0)=-18 y cumple la condicion de x(-n) = x(n)

### 3. Coeficientes de la seria de fourier

```python
#numero de coeficientes a calcular
k=12
# Calcular los coeficientes de la serie de Fourier para números pares (a_k)
a_k = 0
for n in range(-N//2, N//2):
    a_k += x(n, N) * np.cos(2 * np.pi * k * n / N)

a_k *= 1/N

# Calcular los coeficientes de la serie de Fourier para números impares (b_k)
b_k = 0
for n in range(-N//2, N//2):
    b_k += x(n, N) * np.sin(2 * np.pi * k * n / N)

b_k *= 1/N
```

#### 3.1 coeficientes con k=12

```python
def fourier_coefs(x_n,Ncoef):
    # x n: senal definida en un periodo
    # Ncoef: numero de componentes a hallar
    N=len(x_n)
    cn=np.empty(2*Ncoef-1,'complex')
    n=np.arange(0,len(x_n),1)
    
    for k in np.arange(Ncoef):
        cn[k]= (1/N)*np.sum(x_n*np.exp((-2j*np.pi*k*n)/N)) # parte positiva de los coeficientes
        cn[-k]= (1/N)*np.sum(x_n*np.exp((-2j*np.pi*k*n)/N)) #parte negativa de los coeficientes
    return cn
```

```python
cks=12
ck=fourier_coefs(x_values,cks)
plt.figure()
plt.title('Coeficientes')
plt.xlabel('ck(i)')
plt.ylabel('amplitud')
#centrar los cks
ck_centro=abs(np.roll(ck, int(cks-1)))

plt.stem(np.arange(-cks,cks-1,1),ck_centro)
```

#### 3.2 y 3.3) grafica espectro de potencia y su analisis

```python
def calculate_fourier_coefficients(N, k_max):
    a_coefficients = []
    b_coefficients = []
    for k in range(1, k_max + 1):
        a_k = 0
        b_k = 0
        for n in range(-N//2, N//2):
            a_k += x(n, N) * np.cos(2 * np.pi * k * n / N)
            b_k += x(n, N) * np.sin(2 * np.pi * k * n / N)
        a_k *= 1/N
        b_k *= 1/N
        a_coefficients.append(a_k)
        b_coefficients.append(b_k)
    return a_coefficients, b_coefficients

def plot_power_spectrum(a_coefficients, b_coefficients):
    power_spectrum = [a**2 + b**2 for a, b in zip(a_coefficients, b_coefficients)]
    plt.plot(power_spectrum)
    plt.xlabel('k')
    plt.ylabel('Power Spectrum')
    plt.title('Power Spectrum')
    plt.grid(True)
    plt.show()

# Calcular los coeficientes de Fourier
a_coefficients, b_coefficients = calculate_fourier_coefficients(N, k)

# Graficar el espectro de potencia
plot_power_spectrum(a_coefficients, b_coefficients)
```

#### * Al tener comportamiento simetrico par, solo se utiliza la parte positiva para un mejor analisis
#### * En el espectro mostrado en la figura anterior se muestra que conforme a que vamos aumentando la cantidad de coeficientes estos aportan muy poca potencia o casi nula respecto a los componentes iniciales
#### * Se evidencia que alrrededor del 90% de la densisdad espectral de potencia se genera por los primeros 4 coeficientes
#### * los componentes postaeriores al ck4 se muestra que tu contribucion de potencia a la señal es casi nula o nula
#### * Al ser una funcion par como se evidencio anteriormente, los coeficientes de caracter impar no estan contribuyendo potencia al espectro de la señal


#### 3.4) corroborar teorema de parseval

```python
# potencia por teorema de parseval
N=len(a_coefficients)
potencia_senal_parseval = sum(np.abs(a_coefficients)**2+np.abs(b_coefficients)**2)/N
print("Potencia por teorema de parseval:", potencia_senal_parseval)

# potencia teórica
potencia= sum(x**2 for x in [B,A]) / N
print("Potencia teórica:", potencia)
```

```python
error= ((potencia-potencia_senal_parseval)/potencia)*100
error
```

#### como se muestra obtuvimos un error de diferencia alrrededor del 70% de la potencia por teorema de parseval respecto a la potencia original, y esto es debido a que solo tomamos como muestra 12 coeficientes y  para calcularlo de forma precisa se es nesesario casi infinitos coeficientes de fourier para poder igualar al dato real

### 4. Sintesis de señales

#### 4.1) reconstruir señal a partir de cks

```python
def synthezise_fourier(ck,N,NumP):
    #ck: Coeficientes de la serie
    #N: Periodo de la senal
    #Nump: Numero de periodos a graficar
    
    n=np.arange(0,N*NumP,1)
    senalF=np.zeros(N*NumP,'complex')
    
    for k in np.arange(0,len(ck)//2):
        senalF += ck[k]*np.exp((2j*k*np.pi*n)/N) #coeficientes positivos
        senalF += ck[-k]*np.exp((2j*k*np.pi*n)/N) #coeficientes negativos
        if k==0:
            senalF=senalF/2
    return senalF
```

```python
plt.figure()
plt.title('con 12 coeficientes')
plt.xlabel('tiempo')
plt.ylabel('amplitud')
ck_12=fourier_coefs(x_values,12)
ck_cont1=synthezise_fourier(ck_12,N,3)
plt.plot(ck_cont1)
plt.grid()
```

#### 4.2) con k=5 y k=22

#### * k=5

```python
plt.figure()
plt.title('con 5 coeficientes')
plt.xlabel('tiempo')
plt.ylabel('amplitud')
ck_5=fourier_coefs(x_values,5)
ck_cont2=synthezise_fourier(ck_5,N,3)
plt.plot(ck_cont2)
plt.grid()
```

#### * k= 22

```python
plt.figure()
plt.title('con 22 coeficientes')
plt.xlabel('tiempo')
plt.ylabel('amplitud')
ck_22=fourier_coefs(x_values,22)
ck_cont3=synthezise_fourier(ck_22,N,3)
plt.plot(ck_cont3)
plt.grid()
```
